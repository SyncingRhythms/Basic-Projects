```python
import pandas as pd
star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')
star_wars.head(10)
```

```python
star_wars.columns
```

```python
star_wars = star_wars.loc[pd.notnull(star_wars['RespondentID']),:]
```

```python
star_wars.head()
```

```python
yes_no = {
    'Yes': True,
    'No': False,
    float('nan'): False
}
cols_bool = ['Have you seen any of the 6 films in the Star Wars franchise?',
             'Do you consider yourself to be a fan of the Star Wars film franchise?']
for c in cols_bool:
    star_wars[c] = star_wars[c].map(yes_no)
star_wars.head()
```

```python
import numpy as np

film_cols = star_wars.columns[3:9]
col_rename = {}
for i, film in enumerate(film_cols):
    bool_dict = {float('NaN'): False}
    film_key = star_wars[film].unique()
    film_key = film_key[~pd.isnull(film_key)][0]
    bool_dict[film_key] = True
    star_wars[film] = star_wars[film].map(bool_dict)
    col_rename[film] = 'seen' + str(i+1)
star_wars = star_wars.rename(columns = col_rename)
star_wars.head(10)    
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
col_rename = {}
for i, c in enumerate(star_wars.columns[9:15]):
    col_rename[c] = 'ranking' + str(i+1)
star_wars = star_wars.rename(columns=col_rename)
star_wars.head(10)
```

```python
%matplotlib inline

star_wars[star_wars.columns[9:15]].mean().plot.bar()
```

To this point, I've cleaned the data, converting responses to boolean or integers (as appropriate) and renaming ambiguous column names. Then in the cell above, I've plotted the average rankings for each of the 6 Star Wars films. The most liked film (with the lowest ranking) is the 5th film (ranking5; The Empire Strikes Back), as expected, followed by the 6th film.

```python
star_wars[star_wars.columns[3:9]].sum().plot.bar()
```

The most watched of the films is also The Empire Strikes Back (seen5), followed by the 6th film, which correlate nicely with the likability rankings above.

```python
import matplotlib.pyplot as plt
trek_fan_yes = star_wars[star_wars['Do you consider yourself to be a fan of the Star Trek franchise?']=='Yes']
trek_fan_no = star_wars[star_wars['Do you consider yourself to be a fan of the Star Trek franchise?']=='No']

fig = plt.figure(figsize=(10,4))

ax1 = fig.add_subplot(1,2,1)
trek_fan_yes[trek_fan_yes.columns[9:15]].mean().plot.bar(title='Star Trek Fans')
ax2 = fig.add_subplot(1,2,2)
trek_fan_no[trek_fan_no.columns[9:15]].mean().plot.bar(title='Not Star Trek Fans')
```

```python
fig = plt.figure(figsize=(10,4))
ax1 = fig.add_subplot(1,2,1)
trek_fan_yes[trek_fan_yes.columns[3:9]].sum().plot.bar(title='Star Trek Fans', ylim=(0,450))
ax2 = fig.add_subplot(1,2,2)
trek_fan_no[trek_fan_no.columns[3:9]].sum().plot.bar(title="Not Star Trek Fans", ylim=(0,450))
```

Fans and non-fans of Star Trek agree on the rankings of the Star Wars films. However, Star Trek non-fans are less likely to have seen each of the films, with the most robust differences present for films 2, 3, and 4.
